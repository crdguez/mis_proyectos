```python
import matplotlib.pyplot as plt
import numpy as np

# Plot circle or radius 3

an = np.linspace(0, 2*np.pi, 100)

plt.subplot(221)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.title('not equal, looks like ellipse', fontsize=10)

plt.subplot(222)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.title('equal, looks like circle', fontsize=10)

plt.subplot(223)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.axis([-3, 3, -3, 3])
plt.title('looks like circle, even after changing limits', fontsize=10)

plt.subplot(224)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.axis([-3, 3, -3, 3])
plt.plot([0, 4], [0, 4])
plt.title('still equal after adding line', fontsize=10)



plt.show()
```

```python
#https://stackoverflow.com/questions/40747474/sympy-and-plotting

import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
from sympy import *

x, y = symbols("x y")
hp = plot_implicit(Eq(x**2 + y**2, 4), (x, -3, 3), (y, -3, 3))
fig = hp._backend.fig
ax = hp._backend.ax
xx = yy = np.linspace(-3,3)
ax.plot(xx,yy) # y = x
ax.plot([0],[0],'o') # Point (0,0)
ax.set_aspect('equal','datalim')
fig.canvas.draw()
```

```python
from sympy import *
from random import randint

rad_max = 3
m1, m2, b1, b2 = (randint(1,rad_max)), (randint(1,rad_max)), (randint(1,rad_max)), (randint(1,rad_max)) 

M, B = Point(m1, m2), Point(b1, b2)

A = M + (M - B)

print(A, M, B)

print(expand(Line(A,B).equation()))
print(latex(expand(Line(A,B).equation())))
print(expand(Circle(A,M.distance(B)).equation()))

p1 = plot_implicit(Line(A,B).equation(), title='Prueba', xlabel='Prueba2', show=False)
p1.extend(plot_implicit(Circle(M,M.distance(B)).equation(), show=False)) 
p1.extend(plot_implicit(Circle(A,0.1).equation(), show=False)) 
p1.extend(plot_implicit(Circle(B,0.1).equation(),  show=False)) 


%matplotlib notebook
p1.show()
p1._backend.ax.set_aspect('equal')

latex(Line(A,B).equation()), latex(expand(Circle(A,M.distance(B)).equation()))

```
